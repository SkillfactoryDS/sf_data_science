###Задача №1

**Условие задачи.** Написать программу для определения, является ли данное время суток утром, и вывести соответствующее сообщение. Утром считается временной промежуток с 6 (включительно) и до 12 часов (не включительно).
Текущее время находится в переменной now и представлено в формате 'hour:minute:second'. 


**Решение:**

````python
now = '11:24:39'

hour, minute, second = now.split(':')
hour, minute, second = int(hour), int(minute), int(second)

if 6 <= hour < 12:
  print('Morning!')
````

###Задача №2

**Условие задачи.** В математике координатную плоскость условно разделяют на 4 четверти (обычно они обозначаются римскими цифрами:
* Первая четверть (I), в которой находятся точки, у которых и координата x > 0, и координата y > 0 
* Вторая четверть (II), в которой находятся точки, у которых координата x < 0, а координата y > 0
* Третья четверть (III), в которой находятся точки, у которых и координата x < 0 и координата y < 0 
* Четвертая четверть (IV), в которой находятся точки, у которых координата x > 0, а координата y < 0. 

Нам необходимо написать программу, которая определяет, к какой четверти принадлежит точка. Координаты точки, которую нужно проверить хранятся в переменных x и y. Например точка со следующими координатами:

```python
x, y = (6.3, -5.3)
```


принадлежит четвертой четверти (IV) координатной плоскости.



**Решение:**

**Первый способ:**


````python
x, y = (6.3, -5.3)
````

````python
if x > 0:
    if y > 0:                # x > 0, y > 0
        print("I quarter")   # Первая четверть
    else:                    # x > 0, y < 0
        print("IV quarter")  # Четвёртая четверть
else:
    if y > 0:                # x < 0, y > 0
        print("II quarter")  # Вторая четверть
    else:                    # x < 0, y < 0
        print("III quarter") # Третья четверть
````

**Второй способ:**

````python
x, y = (1, 2)
if x > 0 and y > 0:      # x > 0, y > 0
    print("I quarter")   # Первая четверть
if x < 0 and y > 0:      # x < 0, y > 0 
    print("II quarter")  # Вторая четверть
if x < 0 and y < 0:      # x < 0, y < 0
    print("III quarter") # Третья четверть
if x > 0 and y < 0:      # x > 0, y < 0
    print("IV quarter")  # Четвёртая четверть
````

### <center> Конструкция if-elif-else </center>

Общий синтаксис такой конструкции будет следующим:


```python
if Условие 1:
    Блок инструкций 1
elif Условие 2:
    Блок инструкций 2
...
elif Условие n:
    Блок инструкций n
else:
    Блок инструкций n+1
```





Рассмотрим простейший пример: необходимо сравнить значение переменной a с числом 10. Всего есть 3 исхода: число a равно 10, меньше 10 и больше него. Тогда условный оператор будет записан в следующем виде:


````python
a = 16
if a == 10:   # a равно 10  
    print('a equals 10') 
elif a < 10:  # a меньше 10
    print('a is less than 10') 
else:         # a больше 10
    print('a is more than 10') 
````

Благодаря конструкции if-elif мы можем записать решение нашей прошлой задачи о размещении точки на координатной плоскости в более корректном виде:


````python
x, y = (-5, 10)
if x > 0 and y > 0:        # x > 0, y > 0
    print("I quarter")     # Первая четверть
elif x < 0 and y > 0:      # x < 0, y > 0 
    print("II quarter")    # Вторая четверть
elif x < 0 and y < 0:      # x < 0, y < 0
    print("III quarter")   # Третья четверть
elif x > 0 and y < 0:      # x > 0, y < 0
    print("IV quarter")    # Четвёртая четверть
print('Hello')
````

### Задача №3
**Условие задачи.** В переменной month находится номер месяца. Необходимо определить сезон в зависимости от номера месяца и вывести сообщение:
* "Spring" для 3, 4, 5 месяцев;
* "Summer" для 6, 7, 8 месяцев;
* "Autumn" для 9, 10, 11 месяцев;
* "Winter" для 12, 1, 2 месяцев;
* "Incorrect month number" для всех остальных значений.

**Решение:**


````python
month = 5

if month in [3, 4, 5]:
  print('Spring')
elif month in [6, 7, 8]:
  print('Summer')
elif month in [9, 10, 11]:
  print('Autumn')
elif month in [11, 12, 1]:
  print('Winter')
else:
  print('Incorrect month number')
````

### Задача №4

**Условие задачи.** Мы занимаемся разработкой сервиса доставки еды из нашего ресторана. В словаре `dish_time_dict` представлены блюда, которые в данный момент ресторан может приготовить и среднее время их приготовления (в минутах). 

В словаре `street_time_dict` указаны районы города и среднее время доставки в эти районы (в минутах). 




Блюдо заказанное пользователем и район, в котором он живет, хранятся в переменных `dish` и `street` соответственно. Пример:


```python
dish, street = 'Рамен с говядиной', 'Заводской'
```


Необходимо реализовать программу, которая проверяет,  успеет курьер доставить заказ за 60 (включительно) минут с момента его принятия или нет.

Также необходимо реализовать проверку, что блюдо, заказанное клиентом в данный момент, доступно в ресторане, а район, в котором находится клиент доступен для доставки.


````python
dish_time_dict = {
    'Рамен с говядиной': 15,
    'Суши': 18,
    'Лагман с курицей': 20,
    'Лагман с говядиной': 24,
    'Плов с курицей': 28
}
street_time_dict  = {
    'Дзержинский': 39,
    'Солнечный': 40,
    'Заводской': 27,
    'Гагаринский': 43,
    'Кировский': 37,
    'Октябрьский': 34
}

dish, street = 'Лагман с курицей', 'Гагаринский'

if street not in street_time_dict:
    print('Доставка в ваш район недоступна')
elif dish not in dish_time_dict:
    print('Блюдо недоступно, закажите что-то другое')
else:
    dish_time = dish_time_dict[dish]
    street_time = street_time_dict[street]
    full_time = dish_time + street_time 
    delay = full_time - 60
    if delay <= 0:
        print('Заказ будет доставлен вовремя')
    else:
        print('Курьер задержится на {} минут'.format(delay))
````

### Задача №6
**Условие задачи.** Представим ситуацию. В последний месяц удача вам улыбнулась очень плохо. У вас 3 раза взломали пароль от вашего профиля в соц. сети. Вы задумались над тем, что неправильно подходите к вопросу составления паролей. Для этого вы решаете написать программу на Python, которая будет проверять ваш пароль на надежность. Немного погуглив, вы составили следующие рекомендации к паролю:
 
1. Длина – 8 символов (если меньше – то проще взломать, а если длиннее – то сложно запомнить)
2. Пароль не должен состоять только из цифр
3. Пароль не должен состоять только из букв
4. В пароле должны быть заглавные символы
5. В пароле должны быть строчные символы
6. В пароле должен быть хотя бы один из специальных символом ('*', '-', '#'), но каждый из символов должен входить в пароль не более 1 раза.
7. Другие спецсимволы недопустимы, так как вы их не можете запомнить. Вот список запрещенных символов: '!', '@', '$', '%', '^', '`', '&', '(', ')', '+', '=', '_', ‘~'
 
В случае надежного пароля программа должна выводить на печать фразу "Пароль идеален", а в остальных случаях будут перечислены все ошибки, которые вы допустили.
 
**Решение:**


````python
errors_dict = {
    'length': 'Длина пароля не равна 8 символам',
    'digits': 'Пароль состоит только из цифр',
    'letters': 'Пароль состоит только из букв',
    'upper': 'Отсутствуют заглавные буквы',
    'lower': 'Отсутствуют строчные буквы',
    'spec': 'Отсутствуют спецсимволы в пароле',
    'bad_symbols': 'В пароле использованы непредусмотренные символы'
}

bad_symbols = ['!', '@', '$', '%', '^', '`', '&', '(', ')', '+', '=', '_', '~']

password = 'aafaf*al'

# Проверяем длину пароля
if len(password) == 8:
    errors_dict.pop('length')

# Проверяем, что пароль НЕ состоит только из цифр
if not password.isdigit():
    errors_dict.pop('digits')

# Проверяем, что пароль НЕ состоит только из букв
if not password.isalpha():
    errors_dict.pop('letters')

# Проверяем, есть ли в пароле заглавные буквы
if password.lower() != password:
    errors_dict.pop('upper')

# Проверяем, есть ли в пароле строчные буквы
if password.upper() != password:
    errors_dict.pop('lower')

# Проверяем, есть ли в пароле специальные символы
if '*' in password or '-' in password or '#' in password:
    errors_dict.pop('spec')
    # Проверяем, встречаются ли спец символы в пароле несколько раз
    if password.count('*') > 1 or password.count('-') > 1 or password.count('#') > 1:
        errors_dict['spec_count'] = 'Какой-то из спецсимволов в пароле использован более одного раза'

# Проверяем, есть ли в пароле запрещенные символы
if len(set(password).intersection(set(bad_symbols))) == 0:
    errors_dict.pop('bad_symbols')

# Проверяем результирующую длину чек-листа
if len(errors_dict) > 0:
    print(list(errors_dict.values()))
else:
    print('Пароль идеален!')
````
